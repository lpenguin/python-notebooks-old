```python
from math import factorial as fac

def it_shift(items):
    pos = 1
    for item in items:
        yield (item, items[pos:])
        pos +=1
        
def it_pos(items):
    i = 0
    for item in items:
        yield (item, i)
        i += 1

def it_split(items):
    pos = 1
    for item in items:
        yield (item, items[:pos-1] + items[pos:])
        pos +=1
        
def choose(items, count):
    if count == 0:
        yield []     
    
    for (item, items) in it_shift(items):
        for option in choose(items, count - 1):
            yield [item] + option
            
def arrange(items, count):
    if count == 0:
        yield []     
    
    for (item, items) in it_split(items):
        for option in arrange(items, count - 1):
            yield [item] + option
        
def print_sets(sets):
    for (subset, i) in it_pos(sets):
        print "#%d %s" % (i + 1, subset)
        

l = []
for i in xrange(10):
    l.append(i)

def c_choose(n, k):
    return fac(n)/fac(k)/fac(n-k)
# print "arrange"
# print_sets(arrange(['a', 'b', 'c', 'd'], 4))

# print "choose"
# r = range(10)
# k = 5
# print c_choose(len(r), k)
# print_sets(choose(r, k))

r = ['p', 'v', 'k', 's'] + ['a'] * 17
# for option in arrange(r, len(r)):
#     print option
```
