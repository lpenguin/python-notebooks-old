```python
import matplotlib.pyplot as plt
from numpy import *
```

```python
a = 1
x = arange(1, 100, 1)
y = x ** a
l = log2(array(y))

plt.plot(x, y)
plt.grid(True)
plt.title("x - x ** %d" % a)
plt.show()

plt.plot(x, l)
plt.grid(True)
plt.title("x - log2(x ** %d)" % a)
plt.show()

plt.axis([0, 10, 0, 10])
plt.plot(log2(x), l)
plt.grid(True)
plt.title("log2(x) - log2(x ** %d)" % a)
plt.show()
```
